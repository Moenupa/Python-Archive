In <b>lesson 6.1</b>, some undersampling (Near-Miss) and oversampling (SMOTE, ADASYN) techniques use kNN algorithms to consider a neighborhood of points (more representative) at a time instead of just one point. Here we would discuss in details how kNN works.

# Introduction
- kNN is a <b>distance-based supervised</b> ML algorithm.
- It is <b>non-linear</b>, i.e. the decision boundaries generated are not lines or hyperplanes.
- It can be used for <b>both classification and regression</b>.
- It is a <b>lazy learning alogrithm</b> because it generalizes the training data (calculate distance between train and test data points) only when query is made (test data is input).

# Algorithm
The general steps of the kNN algorithm are as follows:
1. Choose an integer k
2. For each point $x$ in test data, calculate the distance between $x$ and all points in the training set. There are many distance metrics to choose from, e.g. Euclidean, Manhattan, Hamming (for categorical data), etc.
3. Consider the k points having the shortest distances with $x$
4. $x$ is assigned to the class appears most frequently to the k points.

# Minimal Examples with sklearn

## Classification

```python
# load iris data
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# split data
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, stratify=y, random_state=42)
```

```python
# knn classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
n_neighbors = 3
knn = KNeighborsClassifier(n_neighbors=n_neighbors)    # set value of k
knn.fit(Xtrain, ytrain)
ypred = knn.predict(Xtest)
accuracy_score(ytest, ypred)
```

```python
%%time
# from sklearn example
# plot decision boundary of kNN in 2D using first 2 features of iris dataset
from matplotlib.colors import ListedColormap

def plot_knn_decision_boundary(X, y, model, data=iris):
    # define colormap for plot
    cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])    # for contonur
    cmap_bold = ['darkorange', 'c', 'darkblue']                          # for data points 

    # generate meshgrid of 2d with step=h=.02
    h = .02                                                              # step of meshgrid
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(
        np.arange(x1_min, x1_max, h), 
        np.arange(x2_min, x2_max, h)
    )
    
    # fit X, y to kNN model and use it to predict all points on meshgrid
    clf = KNeighborsClassifier(n_neighbors=n_neighbors) 
    clf.fit(X[:, :2], y)
    Z = clf.predict(np.c_[xx1.ravel(), xx2.ravel()])                     # put the two dimensions together to form test data points
    Z = Z.reshape(xx1.shape)                                             # reshape to be same as meshgrid
    
    # plot scatter of training data and 
    plt.figure(figsize=(8, 6))
    plt.contourf(xx1, xx2, Z, cmap=cmap_light)
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=data.target_names[y], palette=cmap_bold, alpha=1.0, edgecolor="black")
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    plt.title("3-Class classification showing only first two features(k = %i)"% (n_neighbors));
    plt.xlabel(iris.feature_names[0])
    plt.ylabel(iris.feature_names[1])

# call the function
plot_knn_decision_boundary(X, y, knn);
```

## Regression
- Instead of voting the class for the new data by the k-nearest neighbors, the <b>kNN regressor takes the average target values of the nearest points</b> and assign it to predicted value of the new data point.

```python
boston = datasets.load_boston()
X = boston.data
y = boston.target
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_percentage_error
n_neighbors = 3
knn = KNeighborsRegressor(n_neighbors=n_neighbors, weights='distance')
knn.fit(Xtrain, ytrain)
ypred = knn.predict(Xtest)
mean_absolute_percentage_error(ytest, ypred)    # average abs percent deviation from true value
```
