```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
from IPython import display
import pandas as pd
# pd.options.display.max_rows = 999
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

- In this exercise, we'll investigate a mobile transactions dataset. It is a synthetic dataset generated by a simulator called PaySim which resembles the normal operation of transactions from private data. (https://www.kaggle.com/ealaxi/paysim1)
- We will use logistic regression to classify whether a transaction is fraud or not.

# Feature Definition
| Variable | Definition | Key |
|---|---|---|
| step | maps a unit of time in the real world. In this case 1 step is 1 hour of time. Total steps 744 (31 days simulation) |
| type  | CASH-IN, CASH-OUT, DEBIT, PAYMENT and TRANSFER | 
| amount | amount of the transaction in local currency |	
| nameOrig  | customer who started the transaction |
| oldbalanceOrg | initial balance before the transaction |
| newbalanceOrig | new balance after the transaction |
| nameDest | customer who is the recipient of the transaction |
| oldbalanceDest | initial balance recipient before the transaction. Note that there is not information for customers that start with M (Merchants) |
| newbalanceDest | new balance recipient after the transaction. Note that there is not information for customers that start with M (Merchants) |
| isFraud | This is the transactions made by the fraudulent agents inside the simulation. In this specific dataset the fraudulent behavior of the agents aims to profit by taking control or customers accounts and try to empty the funds by transferring to another account and then cashing out of the system. |
| isFlaggedFraud | The business model aims to control massive transfers from one account to another and flags illegal attempts. An illegal attempt in this dataset is an attempt to transfer more than 200.000 in a single transaction.|

# Preprocessing

## Numrical features
- Logistic regression is sensitive to scale of variables, using sklearn StandardScaler, standardized all numerical feature columns.

```python
numerical_cols = list(df.drop('isFraud', axis=1).select_dtypes(["int64", "float64"]).columns)
numerical_cols
```

```python
# scaling 
from sklearn.preprocessing import StandardScaler
def df_standscaler(df, cols):
    df = df.copy(deep=True)
    scaler = StandardScaler()
    for col in cols:
        df[col] = scaler.fit_transform(df[col].to_numpy().reshape(-1, 1))
    return df
df = df_standscaler(df, numerical_cols)
```

## Categorical features
- Find all categorical feature columns.

```python
# list of categorical columns
cat_cols = list(df.select_dtypes("object").columns)
cat_cols
```

- nameOrig and nameDest are codes that represent clients, for simplicity we extract the first letter as column feature only.
- Check the unique values of each column.

```python
# extract first letter 
df['nameOrig'] = df['nameOrig'].apply(lambda x: x[0])
df['nameDest'] = df['nameDest'].apply(lambda x: x[0])
print(df['nameOrig'].unique())
print(df['nameDest'].unique())
```

- nameOrig only contains customers (C), drop the column.

```python
# drop feature
df.drop(['nameOrig'], axis=1, inplace=True)
df.columns
```

- Using pandas get_dummies(), create one-hot encoding for categorical columns.
- Join the one-hot encoded features to df.
- Delete the old categorical columns.

```python
# one-hot encoding
cat_cols = [x for x in cat_cols if x!='nameOrig']
one_hot = pd.get_dummies(df[cat_cols])
df = df.join(one_hot)
df.drop(cat_cols, axis=1, inplace=True)
df.columns
```

# Train and test logistic model
- Split dataset into train and test sets, fix random_state=42.

```python
# split data
from sklearn.model_selection import train_test_split
X = df.drop(['isFraud'], axis=1)
y = df['isFraud']
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, stratify=y, random_state=42)
```

- Train logistic regression model.
- Predict Xtest.
- Print classification report for test result, the metric we consider here is F1. 

```python
%%time
# logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
lr = LogisticRegression()
lr.fit(Xtrain, ytrain)
ypred = lr.predict(Xtest)
ypred_proba = lr.predict_proba(Xtest)
print(classification_report(ytest, ypred))
```

- Precision of positive class is high but recall is pretty low, it means there is a large proportion of false negative.
- One can plot a confusion matrix to see what happens.

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
def plot_confusion_matrix(y_true, y_pred, labels):
    mat = confusion_matrix(y_true, y_pred, labels=labels)    # Cij = no. of observations belongs to class i predicted to class j
    sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, 
                xticklabels=labels, yticklabels=labels)    # transpose to put true label to x-axis
    plt.xlabel('true label')
    plt.ylabel('predicted label');
plot_confusion_matrix(ytest, ypred, labels=[0, 1])
```

# Over-sampling with SMOTE
- Using SMOTE, over-sample positive class of the training set to increase sensitivity of the model.
- In SMOTE, set random state to 42 and sampling_strategy to 5 times of original ratio at #2

```python
# smote
from imblearn.over_sampling import SMOTE
from collections import Counter
smote = SMOTE(random_state=42, sampling_strategy=5*ratio)
Xtrain_smote, ytrain_smote = smote.fit_resample(Xtrain, ytrain)
print('Resample dataset shape', Counter(ytrain_smote))
```

- Train a new logistic regression model using the resampled training data.
- Print classification report and confusion matrix of test result.

```python
# train and test model
lr.fit(Xtrain_smote, ytrain_smote)
ypred_smote = lr.predict(Xtest)
ypred_proba_smote = lr.predict_proba(Xtest)
print(classification_report(ytest, ypred_smote))
plot_confusion_matrix(ytest, ypred_smote, labels=[0, 1])
```

- The new model is 6% more accurate in terms of f1 score. One can see from the confusion matrix that false negative decreases at the expense of false positive. This results a better balance between precision and recall and hence a 6% higher f1 score.
- One can create a pipeline and tune sampling_strategy as a continuous hyperparameter using random search or bayesian search discussed in Ex 5.2, which will yields even slightly better result.
- Actually logistic regression is not a good model for this dataset (though it gives a good demonstration for resampling). Tree-based models such as random forest, xgboost or lightgbm can give f1 score of over 80% with this dataset, even without resampling.
