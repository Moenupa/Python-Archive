# Introduction
- In this section we are going to apply all we have learn in previous sections to work on the air pollution dataset provided by Kaggle Tabular Playground Series - Apr 2021.
- Link: https://www.kaggle.com/c/tabular-playground-series-apr-2021
- It is a synthetic dataset based on the famous Titantic dataset generated by GAN model (https://github.com/sdv-dev/CTGAN).
- We will use sklearn random forest classifier to handle this problem.

## Feature Definitions
| Variable | Definition | Key |
|---|---|---|
| survival (target) | Survival |	0 = No, 1 = Yes |
| pclass | Ticket class | 1 = 1st, 2 = 2nd, 3 = 3rd |
| sex |	Sex |	
| Age |	Age in years |	
| sibsp	| # of siblings / spouses aboard the Titanic |	
| parch | # of parents / children aboard the Titanic |	
| ticket | Ticket number |	
| fare | Passenger fare |	
| cabin | Cabin number |	
| embarked | Port of Embarkation | C = Cherbourg, Q = Queenstown, S = Southampton |

## Variable Notes
- pclass: A proxy for socio-economic status (SES)
    - 1st = Upper
    - 2nd = Middle
    - 3rd = Lower
- age: Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5

- sibsp: The dataset defines family relations in this way...
    - Sibling = brother, sister, stepbrother, stepsister
    - Spouse = husband, wife (mistresses and fianc√©s were ignored)

- parch: The dataset defines family relations in this way...
    - Parent = mother, father
    - Child = daughter, son, stepdaughter, stepson
    - Some children travelled only with a nanny, therefore parch=0 for them.

## Train and Test Set
- The train set contains 12 columns (including PassengerId and Name) and 100000 rows.
- The test contains another 100000 rows.

## Model Evaluation
- It is a binary classification problem with actual values equal to 0/1. The score is <b>the percentage of passengers the model predicts correctly (accuracy)</b>.

## Cabin 

```python
# Relation between Cabin first letter (which actually represents different deck of ship) and Pclass
# https://en.wikipedia.org/wiki/First-class_facilities_of_the_Titanic, only deck A-G for the real Titanic
def get_cabin_first_letter(x):
    try:
        return x[0]
    except:
        return 'M'    # represent missing cabin
df1 = df.copy()
df1['Deck'] = df1['Cabin'].apply(get_cabin_first_letter)
df2 = df1.groupby(['Deck', 'Pclass'])[['PassengerId']].count()
df3 = df2.groupby(level=0).apply(lambda x: round(100*x / float(x.sum()), 0)).unstack()
df3.columns = df3.columns.droplevel()
df3.plot(kind='bar', stacked=True);
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1));
```

<a id='cabineda'></a>Cabin starts with A to D are mostly for 1st class passengers, while Cabin starts with F, G and T are mostly for 2nd and 3rd class passengers.

```python
# survival rate vs Cabin
df2 = df1.groupby(['Deck'])[['Survived']].mean()
df2['Survived'] = df2['Survived'].apply(lambda x: round(100*x))
df2['Not Survived'] = df2['Survived'].apply(lambda x: 100-x)
df2.plot(kind='bar', stacked=True);
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1));
```

Passengers in Cabin starts with B-G had the highest survival rate (>60%), lowest survival rate at Cabin A (27%).

# Filling Missing Data

```python
df.info()
```

<b>Age, Ticket, Fare, Cabin and Embarked</b> have missing data.

## Age
- To fill in missing ages, the simplest way is to fill in by mean or median. If the dateset has extreme outliers (like 1% of people have yearly income \\$1 billion while others only have about \\$100k, then use median.

```python
# age distribution
df.Age.describe()
```

No outliers in age (outside 1.5 IQR). Either mean or median is good to use. We could also look deeper to Age of different groups like different Sex, Pclass, etc, to determine which median to use to fill NaN.

```python
# check age difference for different Sex and Pclass
df.groupby(['Sex', 'Pclass'])[['Age']].median()
```

So passenger with higher Pclass has higher median age, and median age of female and male are slightly absolute different (3-5).

```python
# fill Age by median of each Sex and Pclass group
df['Age'] = df.groupby(['Sex', 'Pclass'])['Age'].apply(lambda x: x.fillna(x.median()))
df['Age'].isna().any()    # check if all nan are filled
```

## Ticket
- Ticket code can be a number or characters followed by number, so let's split them into two parts.

```python
def split_ticket(x):
    try:
        if x.isdigit():
            return '', x
        char, num = x.split()
        return char, num
    except:
        return '', ''
df['TicketHead'] = df['Ticket'].apply(lambda x: split_ticket(x)[0])
df['TicketTail'] = df['Ticket'].apply(lambda x: split_ticket(x)[1])
df
```

```python
df['TicketHead'].unique()
```

```python
df.groupby('Ticket')[['PassengerId']].count().sort_index().sort_values(by='PassengerId', ascending=False).head(100)
```

```python
df.groupby(['TicketHead', 'Pclass']).count()
```

No obvious pattern can be seen from Ticket, let's just fill NaN by empty string first.

```python
df['Ticket'] = df['Ticket'].fillna('')
df['Ticket'].isna().any()
```

## Fare
- As we saw in <a href='#PclassFare'>EDA</a> that Fare is related to Pclass, we better fill missing Fare according to Pclass. But let's look at the difference of Fare with other features as well.

```python
df.groupby(['Pclass', 'Sex'])[['Fare']].median()
```

Fare of female is more expenisve than that of male in the same Pclass.

```python
df['FamilySize'] = df.apply(lambda row: row['SibSp']+row['Parch'], axis=1)
df.groupby(['Pclass', 'Sex', 'FamilySize'])[['Fare']].median()
```

Individual Fare seems does not related to family size.

```python
# fill Fare by median of each Sex and Pclass group
df['Fare'] = df.groupby(['Sex', 'Pclass'])['Fare'].apply(lambda x: x.fillna(x.median()))
df['Fare'].isna().any()    # check if all nan are filled
```

## Cabin
- ~68% of Cabin data are missing, similar to EDA above we just fill them as M.

```python
df['Cabin'] = df['Cabin'].fillna('M')
df['Deck'] = df['Cabin'].apply(lambda x: x[0])
df['Deck'].isna().any()
```

## Embarked
- Since the number of missing Embarked accounts for 0.25% of data, so we just fill in by mode.

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Embarked'].isna().any()
```

## One-hot Encoding
- Decision tree models implemented by sklearn can only accepts numerical inputs (theoretically decision tree can handle categorical inputs directly), we have to convert categorical data into numeric by <b>one-hot encoding</b> before passing them to the model.
- Categorical features we have: Sex, Surname, TicketHead, TicketTail, Deck, CabinNumber, Embarked

```python
%%time 
onehot = pd.get_dummies(df[['Sex', 'Surname', 'TicketHead', 'Deck', 'Embarked']])
features = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'FamilySize', 'TicketCount', 'TicketTail', 'CabinNumber']].join(onehot)
len(features.columns)
```

# Random Forest Classification

```python
# split data
from sklearn.model_selection import train_test_split
size = 10000    # size of data used
X = features[:size]
y = df['Survived'][:size]
X.shape
```

```python
%%time
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
model = RandomForestClassifier(random_state=0)
scores = cross_val_score(model, X, y, cv=5)
np.mean(scores)
# age continuous .770
# age binned categorical .767
# age, fare binned label encode .765
# size=1000, time 31.8s
# size=5000, time 225s
# size=10000, time 460s
# size=10000 with pca, time 20.9s, acc .736
# size=10000 without pca, time 4.51s, acc .752
# size=100000 without pca, time 76s, acc .751
# size=10000, time 411s, acc .769
```

From experiments the estimated time to run a 5-fold cv could be ~5000s which is more than an hour, so for simplicity we only work on 1/10 of the data here.

# Hyperparameter Tuning
- Perform grid search cv on hyperparameters of decision tree

```python
# get list of parameters of random forest model
from pprint import pprint
pprint(model.get_params())
```

Consider the following hyperparmeters:
- n_estimators = no. of trees in the forest
- max_features = max no. of features to consider when choosing the best split
- max_depth = max no. of levels in each tree
- min_samples_split = min no. of samples required to split an internal node
- min_samples_leaf = min no. of samples required to be a leaf node
- bootstrap = if True, use bootstrap sampling to build a tree (sample with replacement), otherwise use all data for each tree (still has random nature in the model with random subset of features to choose in each node)
- max_samples = if bootstrape = True, the no. of samples draw for each tree

```python
# ~no. of splits per tree if split evenly, reference for max_depth
int(np.log(10000)/np.log(2))
```

```python
%%time
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],                  # default = 100
    'criterion': ['gini', 'entropy'],                # default = gini
    'max_depth': [None, 5, 10, 20],                  # default = None
    'max_features': ['sqrt', 'log2'],                # default = None = sqrt
    'min_samples_leaf': [1, 2, 4],                   # default = 1
    'min_samples_split': [2, 4, 8],                  # default = 2
    'bootstrap': [True, False],                      # default = True
    'max_samples': [None, .3, .6, .9]                # default None, draw X.shape[0] samples for each tree
}
grid = GridSearchCV(model, param_grid, cv=5)
# grid.fit(X, y)
```

```python
# number of parameter set to try
num_parameter_set = [len(x) for x in param_grid.values()]
np.prod(num_parameter_set)
```

As the dataset is not small and the number of sets of parameters is huge, for demonstration we use random search to search over 50 sets of hyparameter.

```python
%%time
from sklearn.model_selection import RandomizedSearchCV
search = RandomizedSearchCV(
    estimator = model, 
    param_distributions = param_grid,
    n_iter = 50,
    random_state=0,
    cv=5
)
search.fit(X, y)
# 20 sets: 195s
# 50 sets: 2400s
```

```python
# best parameters
search.best_params_
```

```python
# mean score of best parameters
search.best_score_
```

Attain an accuracy of 78% after hyperparameter tunning, which gives around 1% improvement. It is expected that grid search would give better improvement if one have more computation power.
